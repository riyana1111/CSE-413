```python
import numpy as np
import matplotlib.pyplot as plt

vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

plt.plot(vector1, label='Vector 1')
plt.plot(vector2, label='Vector 2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Plot of Two Vectors')


plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


matrix = np.random.rand(4, 4)


sns.heatmap(matrix, annot=True, cmap='viridis', cbar=True,
            xticklabels=['Column 1', 'Column 2', 'Column 3', 'Column 4'],
            yticklabels=['Row 1', 'Row 2', 'Row 3', 'Row 4'])


plt.xlabel("Columns")
plt.ylabel("Rows")
plt.title("Heatmap of Random Matrix")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


matrix1 = np.random.rand(4, 4)
matrix2 = np.random.rand(4, 4)


addition = matrix1 + matrix2
subtraction = matrix1 - matrix2
multiplication = np.dot(matrix1, matrix2)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))


axes[0].bar(np.arange(16), addition.flatten())
axes[0].set_title('Addition')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')


axes[1].bar(np.arange(16), subtraction.flatten())
axes[1].set_title('Subtraction')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')


axes[2].bar(np.arange(16), multiplication.flatten())
axes[2].set_title('Multiplication')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
